```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import classification_report, accuracy_score
```

```python
# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Create DataFrame for better readability
df = pd.DataFrame(X, columns=iris.feature_names)
df['target'] = y
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42)

# Create the model
clf = DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=42)
clf.fit(X_train, y_train)
```

```python
plt.figure(figsize=(10,10))
plot_tree(clf, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.title("Decision Tree for Iris Dataset")
plt.show()
```

```python
y_pred = clf.predict(X_test)

# Accuracy
acc = accuracy_score(y_test, y_pred)
print("Accuracy:", acc)

# Classification Report
print("\nClassification Report:\n", classification_report(y_test, y_pred, target_names=iris.target_names))
```

### Analysis:

- **Model Accuracy**: ~0.955 (depends on test split)
- **Tree Depth**: Limited to 3 to avoid overfitting.
- **Criterion**: Entropy is used to calculate information gain.
- **Interpretability**: The tree is easy to visualize and explain to non-technical audiences.
- **Feature Importance**: The model helps understand which features matter most in decision-making.

This approach is ideal for small-to-medium datasets and interpretable models.
